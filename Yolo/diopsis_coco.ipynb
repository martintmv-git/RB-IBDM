```python
import os
import numpy as np
import json
import pandas as pd
from pylabel import importer
```

```python
#Specify path to the coco.json file
path_to_annotations = "labels.json"
#Specify the path to the images (if they are in a different folder than the annotations)
path_to_images = "data"

#Import the dataset into the pylable schema 
dataset = importer.ImportCoco(path_to_annotations, path_to_images=path_to_images, name="diopsis_coco")
```

```python
print(f"Number of images: {dataset.analyze.num_images}")
print(f"Number of classes: {dataset.analyze.num_classes}")
print(f"Classes:{dataset.analyze.classes}")
print(f"Class counts:\n{dataset.analyze.class_counts}")
```

# Stratified Group Shuffle Split

```python
dataset.splitter.StratifiedGroupShuffleSplit(train_pct=.7, test_pct=.1, val_pct=.2, weight=0.01, group_col='img_filename', cat_col='cat_name', batch_size=1)
# dataset.splitter.GroupShuffleSplit(train_pct=.6, val_pct=.2, test_pct=.2)
dataset.analyze.ShowClassSplits()
```
