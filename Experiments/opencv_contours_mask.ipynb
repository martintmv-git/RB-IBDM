```python
import cv2 as cv
from google.colab.patches import cv2_imshow as imshow
```

```python
# clean = cv.imread('/content/drive/MyDrive/background_subtraction/20230714235848_lama.jpg')
clean = cv.imread('/content/drive/MyDrive/background_subtraction/20230714235848.jpg')
```

```python
clean_gray = cv.cvtColor(clean, cv.COLOR_BGR2GRAY)
clean_blurred = cv.GaussianBlur(clean_gray, (3, 3), 0)
clean_edged = cv.Canny(clean_blurred, 0, 70)

# ret, thresh = cv.threshold(clean_gray, 127, 255, 0)
contours, hierarchy = cv.findContours(clean_edged, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
```

```python
imshow(clean_edged)
```

```python
cnt = cv.drawContours(clean, contours, -1, (0,255,0), -1)
imshow(cnt)
```
