```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
digital = cv2.imread('tests/data/Flowers.jpg', cv2.IMREAD_GRAYSCALE)
analog = cv2.imread('tests/data/20230714235848.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
orb = cv2.ORB_create()
```

```python
kp1, des1 = orb.detectAndCompute(digital, None)
kp2, des2 = orb.detectAndCompute(analog, None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(des1, des2)
```

```python
print(len(matches))
```

```python
, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS
```

```python
matching_result = cv2.drawMatches(digital, kp1, analog, kp2, matches, None)
```

```python
plt.imshow(matching_result)
plt.show()
```

```python
cv2.imshow('digital', cv2.drawKeypoints(digital, kp1, None))
```

```python
plt.imshow(cv2.cvtColor(cv2.drawKeypoints(digital, kp1, None), cv2.COLOR_BGR2RGB))
plt.show()
```
