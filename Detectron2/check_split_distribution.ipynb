### Removing duplicated annotations

```python
filtered_train_json = []
md5_list = []

for item in train_json['annotations']:
    md5_result = hashlib.md5(json.dumps(item, separators=(',', ':')).encode("utf-8")).hexdigest()
    if md5_result not in md5_list:
        md5_list.append(md5_result)
        filtered_train_json.append(item)
```

```python
len(filtered_train_json)
```

```python
train_json['annotations'] = filtered_train_json
```

```python
train_ann_list = [ann['id'] for ann in train_json['annotations']]
print(len(train_ann_list))
```

```python
filtered_val_json = []
md5_list = []

for item in val_json['annotations']:
    md5_result = hashlib.md5(json.dumps(item, separators=(',', ':')).encode("utf-8")).hexdigest()
    if md5_result not in md5_list:
        md5_list.append(md5_result)
        filtered_val_json.append(item)

len(filtered_val_json)
val_json['annotations'] = filtered_val_json
val_ann_list = [ann['id'] for ann in val_json['annotations']]
print(len(val_ann_list))
```

```python
# Serializing json
train_json_object = json.dumps(train_json, indent=4)
# Writing to sample.json
with open("coco_data_split_flowers_yolo/train/labels/train2.json", "w") as outfile:
    outfile.write(train_json_object)

# Serializing json
val_json_object = json.dumps(val_json, indent=4)
# Writing to sample.json
with open("coco_data_split_flowers_yolo/val/labels/val2.json", "w") as outfile:
    outfile.write(val_json_object)
```

### Remap the validation dataset to have same category IDs and lists

```python
sync_val_dataset = {
    "images": val_json['images'],
    "annotations": [],
    "categories": train_json['categories'],
}
```

```python
# Transfer annotations with new category IDs
for ann in val_json['annotations']:
  old_cat_id = ann['category_id']
  old_cat_name = next(item['name'] for item in val_json['categories'] if item['id']==old_cat_id)
  new_cat_id = next(item['id'] for item in train_json['categories'] if item['name']==old_cat_name)
  ann['category_id'] = new_cat_id
  sync_val_dataset['annotations'].append(ann)
```
